```python
import matplotlib.pyplot as plt
import math
import numpy
print(plt)
```

```python
# 1. Use matplotlib to plot the following equation:
# You'll need to write the code that generates the x and y points.
# Add an anotation for the point 0, 0, the origin.

# y = x**2 - x + 2
```

```python
x = range(-100, 100)
y = [(n **2 - n + 2) for n in x]

plt.plot(x, y)
plt.title(' $ y = x^2 - x + 2$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.annotate('Origin', xy = (0, 0), xytext = (0, 3000), arrowprops = {})
plt.show
```

```python
# 2.  Create and label 4 separate charts for the following 
# equations (choose a range for x that makes sense):
```

```python
x = range(1, 150)
y = [math.sqrt(n) for n in x]
plt.plot(x,y)
plt.ylim(-5, 50)
plt.text(40, 8, 'y = sqrt(x)', fontsize = 12, color = 'r')
plt.title('$y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$y$')
```

```python
# y = x3
```

```python
x = range(-99,100)
y = [n**3 for n in x]
plt.plot(x,y)
plt.text(1, 3, 'y = x3', fontsize = 12, color = 'b')
plt.xlabel('$x$')
plt.ylabel('$y$')
```

```python
# y = 2x
```

```python
x = range(-10, 11)
y = [2 ** n for n in x]
plt.plot(x,y)
plt.text(1, 3, 'y = 2x', fontsize = 12, color = 'orange')
plt.xlabel('$x$')
plt.ylabel('$y$')
```

```python
# y = 1/(x+1)
```

```python
x = range(0, 100)
y = [(1/(n + 1)) for n in x]

x2 = range(-100, -1)
y2 = [1 / (n + 1) for n in x2]

plt.plot(x,y)
plt.plot(x2, y2)
plt.text(-5, 0, 'y = 1/(x+1)', fontsize = 10, color = 'green')
plt.xlabel('$x$')
plt.ylabel('$y$')
```

```python
# 3. Combine the figures you created in the last step into one large 
# figure with 4 subplots.
```

```python
plt.subplot(2, 2, 1)
x = (range(1, 150))
y = [math.sqrt(n) for n in x]
plt.plot(x,y)
plt.xlabel('$x$')
plt.ylabel('$y$')


plt.subplot(2, 2, 2)
x = range(-99,100)
y = [n**3 for n in x]
plt.plot(x,y)
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.subplot(2, 2, 3)
x = range(-10, 11)
y = [2 ** n for n in x]
plt.plot(x,y)
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.subplot(2, 2, 4)
x = range(0, 100)
y = [(1/(n + 1)) for n in x]
x2 = range(-100, -1)
y2 = [1 / (n + 1) for n in x2]
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.plot(x,y)
plt.plot(x2, y2)
plt.suptitle('4 Subplots')
plt.show()
```
