```python
# 4. Using the chipotle dataset from the previous exercise, 
# create a bar chart that shows the 4 most popular items and 
# the revenue produced by each.
```

```python
def get_db_url(host, user, password):
    url = f'mysql+pymysql://{user}:{password}@{host}/chipotle'
    return url
```

```python
from env import host, user, password
url = get_db_url(host, user, password)
```

```python
orders_df = pd.read_sql('SELECT * FROM orders', url)
orders_df
```

```python
#make item_price a float
orders_df['item_price'] = orders_df['item_price'].str.replace('$','').str.replace(',','').astype('float')
```

```python
orders_df
```

```python
#Isolate the top 4 most popular items
item_quantity= orders_df.groupby('item_name').item_price.sum().nlargest(4)
```

```python
item_quantity
```

```python
top_4= item_quantity.reset_index() #change data into a series
```

```python
#Make a bar plot of item_name vs item_price
sns.barplot(x='item_name', y= 'item_price', data=top_4)
```
