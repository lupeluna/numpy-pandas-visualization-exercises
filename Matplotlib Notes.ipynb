```python
#step 1: Define a figure
plt.figure(figsize = (10, 6))

#Step 2: plot data
#all developers
plt.plot(age, dev_y, color = '#444444', linestyle = '--', label = 'All Devs')

# Add python dev salaries
plt.plot(age, py_dev_y, color = 'g', label = 'Python Devs')

#Add JS salaries
plt.plot(age, js_dev_y, color = 'b', label = 'Javascript')


#Step 3: Customize our plot
#Add Title
plt.title('Median Salary in USA by Age')

# Add x and y axis label
plt.ylabel('Median Salary in USD')
plt.xlabel('Age in Years')

# Add gridlines
plt.grid(True, ls = '--')


#Add annotation
plt.annotate('Underpaid', xy = (25, 38000), xytext = (28, 38000), arrowprops = dict(facecolor = 'gray'))

plt.legend()
```

```python
# Plot a mathematical equation y = x^2

x = range(-50, 51)

y = [n ** 2 for n in x]

plt.scatter(x,y)

plt.title('$y = x^2$', fontsize = 14)

plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$x^2$', fontsize = 14)

plt.xticks([-50, -40, -30, -2020, -10, 0, 10, 20, 30, 40, 50], rotation = 45)

plt.show
```
